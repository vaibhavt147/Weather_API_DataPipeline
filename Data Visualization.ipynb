```python
import pandas as pd
```

```python
df_one = pd.read_csv('one_call.csv')
df_current = pd.read_csv('current.csv')
```

#### The data storage
1. The data from two different api calls is stored in csv format
2. To exctract the data of specific city, bellow functions can be used.

#### Note
* Many of the values from matric wind_gust are unavailable.

```python
def city_df(city,df):
    return df[df['city']==city].reset_index(drop=True)
```

```python
city_df('Mumbai',df_one)
```

```python
city_df('Mumbai',df_current)
```

```python
city_df('New York',df_one)
```

```python
df_test = city_df('New York',df_current)
df_test
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df_test['dt'], df_test['wind_speed'], color = 'b',label='New York')
plt.xlabel('Datetime')
plt.xticks(rotation=90)
plt.ylabel('Wind Speed')
plt.title('Wind Speed time series for New York')
plt.legend()
plt.show()
```

```python
plt.bar(df_test['dt'], df_test['wind_deg'], color = 'y',label='New York')
plt.xlabel('Datetime')
plt.xticks(rotation=90)
plt.ylabel('Wind Degree')
plt.title('Wind Degree time series ')
plt.legend()
plt.show()
```
